```python
# whether in a phone number or an address,
# we will often find unwanted punctuation in the data.
# Let's load some data to see how to address that:
```

```python
# Loading DataFrame with data from csv file

import pandas as pd
df=pd.read_csv(r"C:\Users\Guestuser\Downloads\datasets\datasets\gradedata.csv")
df.head()
```

```python
# To remove the unwanted punctuation, we create a functuation that returns all charracters
# that aren't punctuation, and then we apply that function to our dataframe

# stripping punctuation from the address column
import string
exclude = set(string.punctuation)
# ------------------
print(exclude)
#--------------
def remove_punctuation(x):
    try:
        x = ''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x

df.address = df.address.apply(remove_punctuation)
df.head(5)           
```
